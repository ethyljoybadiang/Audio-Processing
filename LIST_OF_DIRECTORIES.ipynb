```python
import os
import csv

# Specify the directory containing the files
folder_path = '/content/drive/MyDrive/Projects/RNN/ProjectAudio-20230820T141630Z-001/ProjectAudio/audiomentation'

# List all files in the directory
file_names = os.listdir(folder_path)

# Initialize a list to store data from files
all_data = []

# Iterate through each file
for file_name in file_names:
    if file_name.endswith('.wav'):  # Adjust the file extension as needed
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            file_content = file.read()
            all_data.append([file_name, file_content])

# Specify the output CSV file
output_csv_file = '/content/drive/MyDrive/Projects/RNN/output.csv'

# Write data to the CSV file
with open(output_csv_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['File Name', 'File Content'])
    csv_writer.writerows(all_data)

print(f'Data from files has been written to {output_csv_file}')
```
